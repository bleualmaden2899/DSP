Part 2. [100 pts] Prove the equality of the following trigonometric functions using Euler identities. Further prove the equality in Python. Furthermore, plot the left-hand side and right-hand side of the equation and their Euler forms as additional proof.  

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Trigonometric functions

#sin(2θ)
def s2t(t):
    return np.sin(2 * t)

#1/2 * (1 - cos(2θ))
def ct(t):
    return 0.5 * (1 - np.cos(2 * t))

# Euler's formula functions
def es2t(t):
    return (np.exp(2j * t) - 1j * np.exp(-2j * t)) / 2j

def ect(t):
    return 1 - (np.exp(2j * t) + np.exp(-2j * t)) / 4
```

```python
# theta value range
tv = np.linspace(0, 2 * np.pi, 1000)

# values of trigonometric functions
sv = s2t(tv)
cv = ct(tv)

# values of Euler's formula functions
euler_sv = es2t(tv)
euler_cv = ect(tv)
```

```python
# Trigonometric forms plot
plt.figure(figsize=(10, 6))
plt.plot(tv, sv, label='sin(2theta)', color='blue')
plt.plot(tv, cv, label='1/2(1-cos(2theta))', linestyle='--', color='green')
plt.title('Forms of Trigonometric')
plt.xlabel('theta')
plt.legend()
plt.show()
```

```python
# Euler forms plot
plt.figure(figsize=(10, 6))
plt.plot(tv, euler_sv.real, label='Euler sin(2theta)', linestyle='-', color='brown')
plt.plot(tv, euler_cv.real, label='Euler 1/2(1-cos(2theta))', linestyle='--', color='yellow')
plt.title('Forms of Eulers')
plt.xlabel('theta')
plt.legend()
plt.show()
```
